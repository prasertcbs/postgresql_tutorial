# WHERE vs. HAVING
## Server: PostgreSQL
## Table: game
## Author: Prasert Kanawattanachai
## YouTube: https://youtu.be/5S0pIX9CocU

```sql
select * from game order by random() limit 10;
```

```sql
select count(*) from game
```

```sql
select * from game
    where year_of_release = 2016 and platform = 'PS4'
```

```sql
select platform, count(*) no_games
    from game 
    where year_of_release = 2016
    group by platform
```

```sql
select platform, count(*) no_games
    from game 
    where year_of_release = 2016
    group by platform
    having count(*) > 30
    order by no_games desc
```

```sql
select platform, count(*) no_games
    from game 
    -- where year_of_release = 2016
    group by platform, year_of_release
    having count(*) > 30 and year_of_release = 2016
    order by no_games desc
```
